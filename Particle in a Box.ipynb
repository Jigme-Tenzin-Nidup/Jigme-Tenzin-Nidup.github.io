```python
import numpy as np
import matplotlib.pyplot as plt

def particle_in_a_box(n, L, x):
    """
    Calculates the wavefunction for a particle in a 1D box.

    Args:
      n: Quantum number (energy level), an integer greater than 0.
      L: Length of the box.
      x: Array of x-coordinates.

    Returns:
      Array of wavefunction values at x.
    """
    if n <= 0:
        raise ValueError("Quantum number (n) must be greater than 0")
    if L <= 0:
        raise ValueError("Box length (L) must be greater than 0")
    
    return np.sqrt(2/L) * np.sin(n * np.pi * x / L)

def plot_wavefunction(n, L):
    """
    Plots the wavefunction for a given energy level.

    Args:
      n: Quantum number (energy level).
      L: Length of the box.
    """
    x = np.linspace(0, L, 200) # Create x-coordinates
    psi = particle_in_a_box(n, L, x) # Calculate wavefunction values

    plt.plot(x, psi)
    plt.xlabel("Position (x)")
    plt.ylabel("Wavefunction (x)")
    plt.title(f"Particle in a Box: n = {n}")
    plt.grid(True)
    plt.show()

    plt.plot(x, psi**2)
    plt.xlabel("Position (x)")
    plt.ylabel("Probability Distribution Function")
    plt.title(f"Probability distribution for Particle in a Box: n = {n}")
    plt.grid(True)
    plt.show()


if __name__ == "__main__":
    L = 1 # Length of the box
    n = 1 # Quantum number (ground state)
    plot_wavefunction(n, L) # Plot ground state

    n = 2 # First excited state
    plot_wavefunction(n, L) # Plot first excited state
```
